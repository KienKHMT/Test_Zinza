```python
class DonHang:
    def __init__(self, ma_dh, dia_chi, san_pham):
        self.ma_dh = ma_dh
        self.dia_chi = dia_chi
        self.san_pham = san_pham

# Cách 1: Sử dụng danh sách và lặp
def gop_don_hang_danh_sach(danh_sach_don_hang):
    don_hang_gop = []
    dia_chi_da_xu_ly = set()

    for don_hang in danh_sach_don_hang:
        if don_hang.dia_chi not in dia_chi_da_xu_ly:
            # Tìm tất cả đơn hàng có cùng địa chỉ
            don_hang_cung_dia_chi = [dh for dh in danh_sach_don_hang if dh.dia_chi == don_hang.dia_chi]

            # Gộp thông tin
            danh_sach_ma_dh = [dh.ma_dh for dh in don_hang_cung_dia_chi]
            danh_sach_san_pham = []
            for dh in don_hang_cung_dia_chi:
                danh_sach_san_pham.extend(dh.san_pham)

            # Tạo đơn hàng gộp
            don_hang_gop.append(DonHang(
                f"Gop_{'+'.join(danh_sach_ma_dh)}",
                don_hang.dia_chi,
                danh_sach_san_pham
            ))

            dia_chi_da_xu_ly.add(don_hang.dia_chi)

    return don_hang_gop

# Cách 2: Sử dụng dictionary để nhóm đơn hàng
def gop_don_hang_bang_dictionary(danh_sach_don_hang):
    # Tạo dictionary
    nhom = {}

    # Nhóm các đơn hàng nếu địa chỉ giống nhau
    for don_hang in danh_sach_don_hang:
        if don_hang.dia_chi in nhom:
            nhom[don_hang.dia_chi]['danh_sach_ma_dh'].append(don_hang.ma_dh)
            nhom[don_hang.dia_chi]['danh_sach_san_pham'].extend(don_hang.san_pham)
        else:
            nhom[don_hang.dia_chi] = {
                'danh_sach_ma_dh': [don_hang.ma_dh],
                'danh_sach_san_pham': don_hang.san_pham.copy()
            }

    # Tạo danh sách đơn hàng sau khi gộp
    don_hang_gop = []
    for dia_chi, du_lieu in nhom.items():
        don_hang_gop.append(DonHang(
            f"Gop_{'+'.join(du_lieu['danh_sach_ma_dh'])}",
            dia_chi,
            du_lieu['danh_sach_san_pham']
        ))

    return don_hang_gop

# Hàm in đơn hàng
def in_don_hang(danh_sach_don_hang):
    for don_hang in danh_sach_don_hang:
        print(f"Ma DH: {don_hang.ma_dh}")
        print(f"Dia chi: {don_hang.dia_chi}")
        print(f"San pham: {don_hang.san_pham}")
        print()

# Hàm kiểm tra kết quả của cả hai cách tiếp cận
def kiem_tra_gop_don_hang(danh_sach_don_hang, ten_kiem_tra):
    print(f"\n--- Kiem tra {ten_kiem_tra} ---")
    # Test cách 1
    print("\nKiem tra cach su dung danh sach:")
    don_hang_gop_danh_sach = gop_don_hang_danh_sach(danh_sach_don_hang)
    in_don_hang(don_hang_gop_danh_sach)
    # Test cách 2
    print("\nKiem tra cach su dung dictionary:")
    don_hang_gop_tu_dien = gop_don_hang_bang_dictionary(danh_sach_don_hang)
    in_don_hang(don_hang_gop_tu_dien)

# Test case 1: Các đơn hàng có địa chỉ trùng nhau
def kiem_tra_cung_dia_chi():
    danh_sach_don_hang = [
        DonHang("DH1", "Hanoi", ["Banh", "Keo"]),
        DonHang("DH2", "Hanoi", ["Nuoc ngot"]),
        DonHang("DH3", "Hanoi", ["Banh"])
    ]
    kiem_tra_gop_don_hang(danh_sach_don_hang, "Cung dia chi")

# Test case 2: Các đơn hàng có địa chỉ khác nhau
def kiem_tra_khac_dia_chi():
    danh_sach_don_hang = [
        DonHang("DH1", "Hanoi", ["Banh", "Keo"]),
        DonHang("DH2", "HCM", ["Bim bim"]),
        DonHang("DH3", "SonLa", ["Thuoc la"])
    ]
    kiem_tra_gop_don_hang(danh_sach_don_hang, "Khac dia chi")

# Test case 3: Trường hợp hỗn hợp
def kiem_tra_hon_hop_dia_chi():
    danh_sach_don_hang = [
        DonHang("DH1", "Hanoi", ["Banh", "Keo"]),
        DonHang("DH2", "HCM", ["Bim bim"]),
        DonHang("DH3", "Hanoi", ["Nuoc ngot"]),
        DonHang("DH4", "SonLa", ["Thuoc la"]),
        DonHang("DH5", "Hanoi", ["Banh"])
    ]
    kiem_tra_gop_don_hang(danh_sach_don_hang, "Tron hon hop dia chi")

# Chạy các test case
if __name__ == "__main__":
    kiem_tra_cung_dia_chi()
    kiem_tra_khac_dia_chi()
    kiem_tra_hon_hop_dia_chi()
```
